```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('/content/drive/MyDrive/star_classification.csv')
```

```python
df.info()
```

```python
df.head()
```

```python
df = df.drop(["obj_ID","run_ID","rerun_ID","cam_col","field_ID","spec_obj_ID","plate"],axis=1)
df = df[df["class"]!= "GALAXY"]
```

```python
from sklearn.preprocessing import LabelEncoder
LE = LabelEncoder()
df["class"] = LE.fit_transform(df["class"])
```

```python
X_train,X_test,y_train,y_test = train_test_split(df.drop("class",axis=1),df["class"],test_size=0.2)
```

```python
Dt = DecisionTreeClassifier()
Dt.fit(X_train , y_train)
```

```python
from sklearn.tree import plot_tree
plot_tree(Dt, filled=True ,max_depth=5)
```

```python
for value in LE.inverse_transform(Dt.predict(X_test)):
  print(value)
```
