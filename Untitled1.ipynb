```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/drive/MyDrive/Used_Bikes.csv")
```

```python
numeric_df = df.select_dtypes(include=['int64', 'float64'])
```

```python
corr = numeric_df.corr()
```

```python
target = "price"
```

```python
if target in corr.columns:
    best_feature = corr[target].drop(target).abs().idxmax()
    print("Best variable for LM Plot with price:", best_feature)
else:
    raise ValueError("Column 'price' not found in dataset")
```

```python
sns.lmplot(x=best_feature, y="price", data=df, height=6, aspect=1.4)
plt.title(f"Best LM Plot: {best_feature} vs Price")
plt.xlabel(best_feature)
plt.ylabel("Price")
plt.show()
```
